# Aplikasi Metode Numerik
Jalankan sel di bawah ini untuk memulai aplikasi interaktif.

```python
from methods.regula_falsi import regula_falsi
from methods.newton_raphson import newton_raphson
from methods.iterasi import iterasi
from utils.display import print_table

def main():
    while True:
        print("\n=== MENU METODE NUMERIK ===")
        print("1. Metode Regula-Falsi")
        print("2. Metode Newton-Raphson")
        print("3. Metode Iterasi Sederhana")
        print("4. Keluar")
        pilihan = input("Pilih menu (1-4): ").strip()
        
        if pilihan == '1':
            print("\n-- Regula-Falsi --")
            f = input("f(x) (misal: x**2 - 4): ")
            a = float(input("Batas bawah (a): "))
            b = float(input("Batas atas (b): "))
            try:
                res = regula_falsi(f, a, b)
                print_table(res)
            except Exception as e:
                print("Error:", e)
        elif pilihan == '2':
            print("\n-- Newton-Raphson --")
            f = input("f(x) (misal: x**2 - 4): ")
            df = input("f'(x) (misal: 2*x): ")
            x0 = float(input("Tebakan awal (x0): "))
            try:
                res = newton_raphson(f, df, x0)
                print_table(res)
            except Exception as e:
                print("Error:", e)
        elif pilihan == '3':
            print("\n-- Iterasi Sederhana --")
            g = input("g(x) bentuk x=g(x) (misal: (x+2)**0.5): ")
            x0 = float(input("Tebakan awal (x0): "))
            try:
                res = iterasi(g, x0)
                print_table(res)
            except Exception as e:
                print("Error:", e)
        elif pilihan == '4':
            print("Terima kasih.")
            break
        else:
            print("Pilihan tidak valid.")

main()
```
